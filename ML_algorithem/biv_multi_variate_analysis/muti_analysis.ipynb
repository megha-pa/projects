```python
tips=sns.load_dataset("tips")
tips
```

```python
iris=sns.load_dataset("iris")
iris
```

## 1.Scatterplot(Numeric-Numeric)

```python

sns.set(rc = {'figure.figsize':(15,8)}) 
sns.scatterplot(tips["total_bill"],tips["tip"],hue=tips["sex"],style=tips["smoker"],size=tips["size"])
```

## 2.Bar Plot(Numerical-Categorical)

```python
sns.barplot(titanic["Pclass"],titanic["Fare"],hue=titanic["Sex"])
```

## 3.Box Plot(Numerical-Categorical)

```python
sns.boxplot(titanic["Sex"],titanic["Age"],hue=titanic["Survived"])
```

## 4.Distplot(Numerical-Categorical)

```python
sns.distplot(titanic[titanic["Survived"]==0]["Age"],hist=False,color="red")
sns.distplot(titanic[titanic["Survived"]==1]["Age"],hist=False,color="green")
```

## 5.Heatmap (Categorical-Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic["Survived"]))
```

```python
(titanic.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

## 6.Clustermap (Categorical-Categorical)

```python
pd.crosstab(titanic["SibSp"],titanic["Survived"])
```

```python
sns.clustermap(pd.crosstab(titanic["SibSp"],titanic["Survived"]))
```
