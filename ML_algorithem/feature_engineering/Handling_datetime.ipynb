```python
## working with data and times
#Link=https://github.com/campusx-official/100-days-of-machine-learning/blob/main/day34-handling-date-and-time/working-with-dates-and-time.ipynb
import numpy as np
import pandas as pd
```

```python
date=pd.read_csv("/Users/pluto7/Desktop/git_data/ML_algorithem/Auto_arima/algo_df.csv")
```

```python
date.info()

```

```python
# Converting to datetime datatype
date['Order Date'] = pd.to_datetime(date['Order Date'])
date.rename(columns = {'Order Date':'date'}, inplace = True)
```

## 1.Extract Year

```python
date['date_year'] = date['date'].dt.year

date.sample(5)
```

## 2.Extract Month

```python
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

## 3.Extract Days

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
## day of week Monday assumed as day of week and denoted as 0
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()
```

```python
date
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

```

```python
date['date_week'] = date['date'].dt.week
```

```python
date['quarter'] = date['date'].dt.quarter
```

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)
```

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
np.round((today -date['date']) / np.timedelta64(1, 'M'),0)
```
